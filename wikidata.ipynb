```python
import pandas as pd 
```

```python
data = pd.read_excel('wikipedia_dataset.xlsx')
data
```

```python
wiki_1 = pd.melt(data, id_vars='Page', 
                  var_name= 'Dates',
                  value_name= 'Daily_Page_Visit_Count'
                  )
wiki_1
```

```python
wiki_2 = wiki_1.rename(columns = {'Page' : 'Search_Topics'})
wiki_2.head()
```

```python
wiki_2.info()
```

```python
wiki_2.dropna(inplace = True)
wiki_2.info()
```

```python
wiki_tidy = (wiki_2

 .assign(Month = lambda x: x['Dates']
        .dt.month_name())

.assign(Day = lambda x: x['Dates']
        .dt.day)       
)

wiki_tidy.head()
```

```python
wiki_tidy.info()
```

SUBSETTING

```python
Jan_wiki = wiki_tidy[wiki_tidy['Month'] == 'January']
Jan_wiki
```

```python
Jan_1 = (Jan_wiki[Jan_wiki['Day']==1])
Jan_1
```

```python
Jan = pd.DataFrame(Jan_1.groupby(["Daily_Page_Visit_Count"])["Search_Topics"])
Jan
```

```python
df_Jan1 = (Jan.rename(columns = {0: "Daily_Page_Visit_Count", 1: 'Search_Topics'}))
df_Jan1
```

MOST TRENDING IN JANUARY

```python
df_Jan1.tail(10)
```
